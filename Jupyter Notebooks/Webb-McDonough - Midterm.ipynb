Import MongoDB

```python
import json
import pymongo
import certifi

with open('C:\\Users\\webbr\\OneDrive\\Documents\\GitHub\\Secrets.json', 'r') as pw:
    data = json.load(pw)

    key = data["examplemongodb"]


print(f"My string is {len(key)} characters long")

ConnectionString = data["examplemongodb"]

client = pymongo.MongoClient(ConnectionString, tlsCAFile=certifi.where())

db = client.MovieDatabase

AllCollections = db.list_collection_names()

print(f"Using MongoDB version {client.server_info()['version']}")
print(f"This Database has the collections {AllCollections}")
```

Get All IMDB Movies From 2009

```python
import pandas as pd 
import re

db.IMDB.release_date = pd.to_datetime(db.IMDB.release_date, errors="coerce")

IMDB_Year = pd.DataFrame(db.IMDB.find({"release_date": re.compile("2009")}))
```

Get All MetaCritic Movies From 2009

```python
import pandas as pd 
import re

db.MetaCritic.release_date = pd.to_datetime(db.MetaCritic.release_date, errors="coerce")

MetaCritic_Year = pd.DataFrame(db.IMDB.find({"release_date": re.compile("2009")}))
MetaCritic_Year.head()
```

Merge the Different Lists

```python
import pandas as pd
unified_view = pd.merge(IMDB_Year, MetaCritic_Year, how="inner", on="title")
unified_view.head()
```
